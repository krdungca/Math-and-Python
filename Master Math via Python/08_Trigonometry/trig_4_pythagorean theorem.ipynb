$a^2 + b^2 = c^2$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 3 
b = 4

c = np.sqrt(a**2 + b**2)
c

plt.plot([0,a],[0,0])
plt.plot([0,0],[0,b])
plt.plot([0,a],[b,0])

plt.axis('square')
plt.axis()
axlim = np.max((a,b)) + 0.5
plt.axis([-2,axlim,-0.5, axlim])

plt.text(a/3,-0.25,'a=' + str(a),fontsize=15)
plt.text(-0.75,b/3,'b=' + str(b),fontsize=15)

plt.show()
```

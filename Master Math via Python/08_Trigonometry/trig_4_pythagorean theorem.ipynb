$a^2 + b^2 = c^2$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 9
b = 15

c = np.sqrt(a**2 + b**2)
c

plt.plot([0,a],[0,0])
plt.plot([0,0],[0,b])
plt.plot([0,a],[b,0])

plt.axis('square')
plt.axis()
axlim = np.max((a,b)) + 0.5
plt.axis([-2,axlim,-0.5, axlim])

plt.text(a/3,-0.25,'a=' + str(a),fontsize=15)
plt.text(0.1,b/3,'b=' + str(b),fontsize=15)
plt.text(a/2+0.1,b/2,'c=' + str(c),fontsize=15)

plt.show()
```

```python
#plot z = 2 + 3j
a = -1
b = 2.5

z = a + b*1j
x = np.real(z)
y = np.imag(z)

plt.plot([0,x],[0,y],'o-')
plt.axis('square')
plt.axis([-4,4,-4,4])

axlim = plt.gca()
x_lim = axlim.get_xlim()
y_lim = axlim.get_ylim()
plt.plot(x_lim,[0,0],'k--',alpha=0.2)
plt.plot([0,0],y_lim,'k--',alpha=0.2)

z = np.sqrt(x**2+y**2)

plt.text(x/2+0.2,y/2,'c=' + str(np.round(z,2)))
plt.show()
```

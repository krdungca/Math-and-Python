Euler's formula/ Euler's identity

"leads to the most elegant equation in all of human civilization"

k = angle

$e^{ik} = cos(k) + i \times sin(k)$

$me^{ik} = m(cos(k) + i \times sin(k))$




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = np.pi / 6
m = 2.3

eul = m * np.exp(1j*k)
cis = m*(np.cos(k) + 1j*np.sin(k))  #cos imag sin = cis
plt.plot([0,np.real(eul)],[0,np.imag(eul)])
```

```python
mag = abs(eul) #distance from origin
ang = np.angle(eul)

print(mag)
print(ang)

plt.polar([0,ang],[0,mag],'r')
plt.polar(k,m,'bo')

#plotting custom thetagrids
plt.thetagrids((350,149,199),labels=None)
plt.thetagrids(range(0,350,9))
plt.show()
```

```python
#create function named eulerFromCosSin()
#input = cosine part, sine part
#generate polar plot
#top label: me^i*phi    m=[magnitude]     , phi = [angle]
#discover how to display custom theta grids



def eulerFromCosSin(cosval,sinval):
    #calculate mag, k, and cis
    mag = abs(np.sqrt(cosval**2 + sinval**2))
    kcos = np.arccos(cosval/mag)
    ksin = np.arcsin(sinval/mag)
    
    cis = mag*(np.cos(kcos) + 1j*np.sin(ksin))

    ang = np.angle(cis)

    #plot polar
    plt.polar([0,ang],[0,mag],'b-')

    #display labels
    plt.title('$me^{i \\phi}, m = %s, \\phi = %s$ or %s°' %(np.round(mag,2),np.round(ang,2),np.round(np.rad2deg(ang),2)))
    
```

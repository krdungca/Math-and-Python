$cos^2 + sin^2 = 1$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,6*np.pi, 100)

plt.plot(x,np.sin(x),'r-',label='$\\sin(\\theta)$')
plt.plot(x,np.cos(x),'b-',label='$\\cos(\\theta)$')

plt.xlabel('rad')
plt.ylabel('fvalue')

plt.legend()
plt.show()
```

```python
th = np.linspace(0,4*np.pi,1000)

plt.plot(th,np.tan(th),'k')
```

Trig identities

sin (x + or - y) = sin_x * cos_y + cos_x * sin_y

cos (x + or - y) = cos_x * cos_y + sin_x * sin_y

sin (2*x) = 2 * sin_x * cos_x

cos (2*x) = cos_x^2 - sin_x^2


Additional info: https://www.geeksforgeeks.org/trigonometric-identities/

```python
#plot the ff:
# y1 = sin( theta + cos(theta)     )
# y2 = cos( theta + sin(theta)     )


theta = np.linspace(0,6*np.pi,500)

y1 = np.sin(theta + np.cos(theta))
y2 = np.cos(theta + np.sin(theta))

plt.plot(theta,y1,label='$y_1 = sin(\\theta + cos(\\theta))$')
plt.plot(theta,y2,label='$y_2 = sin(\\theta + cos(\\theta))$')

plt.legend()
plt.show()
```

```python
#user input value angle (AS DEGREES!)
inputval = int(input('input angle as degrees'))
plt.title('$\\theta = %s$ '%inputval)

#plot the plane
plt.axis([-1.15,1.15,-1.15,1.15])
plt.axis('square')
axlim = plt.gca()
x_lim = axlim.get_xlim()
y_lim = axlim.get_ylim()
plt.plot(x_lim,[0,0],'k--',alpha=0.2)
plt.plot([0,0],y_lim,'k--',alpha=0.2)

#generate a base_circle
x_circle = np.linspace(0,2*np.pi,100)
plt.plot(np.cos(x_circle),np.sin(x_circle),'k-',alpha=0.2)

#angle value as theta (radians)
theta = np.deg2rad(inputval)
plt.plot(np.cos(theta),np.sin(theta),'ko')

#plot arc to theta value
x_arc = np.linspace(0,theta,100)
plt.plot(np.cos(x_arc)/5,np.sin(x_arc)/5,'k-',linewidth=1)
plt.plot(np.cos(x_arc),np.sin(x_arc),'k-',linewidth=3)

#plot sin part(red) 
plt.plot([0,0],[0,np.sin(theta)],'r-',label='sin part',linewidth=2)
plt.plot([0,np.cos(theta)],[np.sin(theta),np.sin(theta)],'r--',alpha=0.5)

#plot cos part(green)
plt.plot([0,np.cos(theta)],[0,0],'g-',label='cos part',linewidth=2)
plt.plot([np.cos(theta),np.cos(theta)],[0,np.sin(theta)],'g--',alpha=0.5)

#plot diagonal from origin to theta point
plt.plot([0,np.cos(theta)],[0,np.sin(theta)],'k--',alpha=0.2)

#final display
plt.legend()
plt.show()
```

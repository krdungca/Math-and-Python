Plot unit vectors (length of 1) with random phase angles.

Thus from the unit circle exercise!

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#cos(x) -> x
#sin(x) -> y
n = 250

#plot circle
whole_circle = 2*np.pi
x = np.linspace(0,whole_circle,100) #2*pi is the value of angle of the whole circle (360degrees)
plt.plot(np.cos(x),np.sin(x),'k')
plt.plot(0,0,'ko')


#generate random values
minval = 0
maxval = whole_circle
rand_angles = np.random.rand(n)
rand_angles = rand_angles * (maxval-minval) + minval

for i in range(0,n):
    plt.plot([0,np.cos(rand_angles[i])],[0,np.sin(rand_angles[i])])

plt.axis('square')
plt.show()
```

```python
n = 50

angs_norm = np.random.randn(n)
angs_rand = np.random.rand(n)
angs_whole = np.random.rand(n)*2*np.pi

plt.subplot(131)
for i in angs_norm:
    plt.plot([0,np.cos(i)],[0,np.sin(i)])

plt.title('Random-Normal')
plt.axis('off')
plt.axis('square')



plt.subplot(132)
for j in angs_rand:
    plt.plot([0,np.cos(j)],[0,np.sin(j)])

plt.title('Random only')
plt.axis('off')
plt.axis('square')


plt.subplot(133)
for k in angs_whole:
    plt.plot([0,np.cos(k)],[0,np.sin(k)])

plt.title('Random to 2*pi')
plt.axis('off')
plt.axis('square')



plt.show()

```

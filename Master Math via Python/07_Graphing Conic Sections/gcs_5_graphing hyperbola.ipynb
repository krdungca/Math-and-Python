Hyperbola

REVERSE OF ELLIPSE! same function but instead of add, it is subtracted

$\frac{(y-k)^2}{b^2} - \frac{(x-h)^2}{a^2} - 1 = 0$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1       #stretching
b = 0.5
h = 1       #shifting
k = 2

axlim = 2*np.max((a,b)) + np.max((abs(h),abs(k)))

x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

#create the function

hyperbola = (((X-h)**2)/a**2) - (((Y-k)**2)/b**2) - 1

plt.plot(h,k,'go')

plt.contour(X,Y,hyperbola,0)

plt.plot([-axlim,axlim],[0,0],'k--')
plt.plot([0,0],[-axlim,axlim],'r--')

plt.gca().set_aspect('equal')
```

```python
n = 20
h = 0
k = 0
a = np.linspace(0,4,n)
b = np.linspace(0,4,n)


x = np.linspace(-8,8,100)
X,Y = np.meshgrid(x,x)

for i in range(0,n):
    F = (X-h)**2/a[i]**2 - (Y-k)**2/b[i]**2 - 1
    plt.contour(X,Y,F,0,colors=[i/n,0,i/n])
    plt.contour(Y,X,F,0,colors=[0,0,i/n])

plt.gca().set_aspect('equal')
plt.axis([-6,6,-6,6])
plt.axis('off')
```

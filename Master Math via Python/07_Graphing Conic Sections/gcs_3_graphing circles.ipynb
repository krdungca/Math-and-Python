Circle Formula

$(y-k)^2 + (x-h)^2 = r^2$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 2
k = -3
r = 5
axlim = r + np.max((abs(h),abs(k)))  #setting the maximum limit-point of the circle(based from which is larger b/w h or k (+r))
                                    #abs() is used to determine the farthest point from the origin

x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

Fxy = (X-h)**2 + (Y-k)**2 - r**2 #set to 0

plt.contour(X,Y,Fxy,0)  #lines of same regions with same height and magnitude
plt.plot([-axlim,axlim],[0,0],'k--') #plotting the base axes
plt.plot([0,0],[-axlim,axlim],'k--')
plt.plot(h,k,'go')

plt.gca().set_aspect('equal')
```

```python
h = [-1.5,1.5]

x = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,x)

for r in np.linspace(0.5,3,15):
    for hi in h:
        Fxy = (X-hi)**2 + (Y)**2 - r**2
        plt.contour(X,Y,Fxy,0,colors=[r/3,r/3,r/3])   #loop of only the 1st circle of the contour

plt.plot(h,[0,0], linewidth=3)

plt.axis('off')
plt.gca().set_aspect('equal')
```

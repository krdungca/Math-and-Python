Parabola

$x^2$ positioned differently

$4p(y-k) = (x-h)^2$

p = parameters

$y = \frac{1}{4p} (x-h)^2 + k$


$vertex : (x,y) \rightarrow (h,k)$ : minimum or maximum point in the parabola 

$focus : (h,k+p)$

$directrix : y = k-p$

Parabola are set of points that is equidistant b/w focus and directrix.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters

a = 1  #a = 1/(4*p)
h = 0
k = 0
n = 100

x = np.linspace(h-5,h+5,n)

y = a*(x-h)**2 + k

plt.grid()


#opens up top/bot
plt.plot(x,y)


#opens up left/right (modified)
plt.plot(y,x-5)

```

```python
#parameters

a = 1  #a = 1/(4*p)
h = 2
k = 2
n = 100

x = np.linspace(h-5,h+5,n)

y = a*(x-h)**2 + k

plt.grid()

#graph parabola
plt.plot(x,y,label='parabola')

#graph vertex (h,k)
plt.plot(h,k,'ro',label='vertex')

#graph focus (h, k+p)
plt.plot(h,k+a,'go',label='focus')

#graph directrix y = k - p
y_d = k - a
# plt.plot([x[0],x[-1]],[y_d,y_d],'r-',label='directrix')
plt.plot(x[[0,-1]],[y_d,y_d],'r-',label='directrix') #iterable, x[  [list that will be iterable] ]
plt.plot(x[[0,58]],[y_d-1,y_d-1],'o-',label='sub_fake_directrix') #iterable, x[  [list that will be iterable] ]

#display legend
plt.legend()

#    y= mx + b   x = (y-b)/m
```

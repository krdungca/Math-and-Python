```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,Y = np.meshgrid(range(0,10),range(0,15))

plt.subplot(121)
plt.pcolormesh(X,edgecolor='k',linewidth=0.1)

plt.subplot(122)
plt.pcolormesh(Y,edgecolor='k',linewidth=0.1)

plt.subplots_adjust(wspace=0.5)


```

```python
x = np.linspace(-np.pi,2*np.pi,40)
y = np.linspace(0,4*np.pi,72)

X,Y = np.meshgrid(x,y)

Fxy = np.cos(X) + np.sin(Y)


plt.imshow(Fxy)

```

```python
#2D gaussian blur

# Z = exp (-(   (X^2 / Y^2)/ s  ))       ), s = kernel size

# ORIGINAL GAUSSIAN FXN (1D) y = e**( -4*ln(2)*t**2      / h**2  )

spaces = 500

t = np.linspace(-1,1,spaces)

h = spaces
exp_num = -4 * np.log(2) * t**2
exp_den = h**2

y = np.exp(exp_num/exp_den)

# plt.plot(t,y)
# plt.axis([-3,3,0,1.1])

# t = 'x-axis'
# y = function
# h = full width at half maximum

X = np.zeros([h,len(t)])
Y = np.zeros([len(t),h])

for i in range(h):
    Y[i,:] = np.exp(-4 * np.log(2) * t**2/h**2)
    X[:,i] = np.exp(-4 * np.log(2) * t**2/h**2)


s = 0.5
z_val = np.exp(((X**2 + Y**2))/s)

# plt.pcolormesh(z_val)

plt.imshow(z_val)
```

```python
x = np.linspace(-2,2,100)
X,Y = np.meshgrid(x,x)

s = 1 #standard deviation

gaus2d = np.exp((-(X**2 + Y**2))/s)

plt.imshow(gaus2d)
plt.axis('off')
```

```python
#complex roots of unity

#   z is an element of complex numbers
#  z^n = 1

# i = sqrt(-1)
# z  = exp ((2 * pi * i * k ) / n)
# k = 0, 1, ..., n-1

#for example

#z^5 = 1... this means there are 5 roots that will equal to 1

n = 150

x = np.linspace(0,2*np.pi,n)

for k in range(0,n):
    z  = k*(np.exp((2 * np.pi * 1j * k ) / n))
    z_line = (k+1)*(np.exp((2 * np.pi * 1j * (k+1) ) / n))
    
    plt.plot([0,np.real(z)],[0,np.imag(z)],color='gray')
    plt.plot([np.real(z),np.real(z_line)],[np.imag(z),np.imag(z_line)],'k')
    
print(k)    
z_last  = (k+1)*(np.exp((2 * np.pi * 1j * (k+1) ) / n))
plt.plot([0,np.real(z_last)],[0,np.imag(z_last)],'k')
plt.axis('square')

axis = plt.gca()

# axis.set_xlim([-2,2])
# axis.set_ylim([-2,2])
```

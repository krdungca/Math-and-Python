```python
#discrete difference operator

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,3,4]

x[1]-x[0]
```

```python
np.diff(x)   #3-1 (2nd - 1st), 4-3 (3rd - 2nd)
```

```python
v = np.arange(0,10)

print(np.diff(v)) #first order
print(np.diff(v,2)) #2nd order   #np.diff( VALUE, ORDER OF DERIVATIVE )
```

```python
#graph f(x) = x**2
x = np.linspace(-2,2,2100)
fx = x**2

plt.xlim(-2,2)
plt.ylim(-1,2)
plt.grid()
plt.plot(x,fx)
plt.axis([-2,2,-1,2])#set the axis    [xstart,xend,ystart,yend]

#graph df / dx
df = np.diff(fx)
dx = np.diff(x)

dfdx = df/dx

plt.plot(x[:-1],dfdx)   #x[:-1] remove the last OR from start until the 2nd to the last 

plt.plot(x[:-1],df)
```

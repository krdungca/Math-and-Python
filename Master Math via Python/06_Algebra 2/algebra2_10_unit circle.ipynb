```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,5000)

plt.plot(np.cos(x),np.sin(x),'k')
plt.axis('square')

#base number-line
axis = plt.gca()
plt.plot([0,0],[axis.get_xlim()[0],axis.get_xlim()[1]],'--',color=[0.8,0.8,0.8])
plt.plot([axis.get_ylim()[0],axis.get_ylim()[1]],[0,0],'--',color=[0.8,0.8,0.8])

angle = np.pi / 4

plt.plot([0,np.cos(angle)],[0,np.sin(angle)]) #review... cos(a) = x   sin(a) = y

```

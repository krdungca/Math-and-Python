```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-3,3,1001)

h = 1

exp_num = -4 * np.log(2) * t**2
exp_den = h**2

y = np.exp(exp_num/exp_den)
# t = 'x-axis'
# y = function
# h = full width at half maximum


plt.plot(t,y)
plt.axis([-3,3,0,1.1])
```

```python

t = np.linspace(-3,3,1001)


N = 100

h = np.zeros(N)
pnts = N

G = np.zeros((N,len(t)))

for i in range(N):
    h[i] = (i+1)/N
    G[i,:] = np.exp( -4*np.log(2)*t**2 / h[i]**2      )


plt.pcolormesh(t,h,G)


```

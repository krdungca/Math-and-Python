```python
# y = e^x    semi-'quadratic'... as x -> -oo  ,y nears 0,,,, as x--> +oo, y = +oo
# y = log x  semi also... reversed   as x ->, y = -oo,   as x--> +oo, y --> 1

#log (-a)  .. will result in a complex number
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
#y=e**x
x = np.linspace(-3,3,99)

e_x = np.exp(x)
plt.subplot(1,2,1)
plt.plot(x,e_x)

log_x = np.log(x)
plt.subplot(1,2,2)
plt.plot(x,log_x)
```

```python
x = np.linspace(0.0001,10,10)

y1 = np.log(np.exp(x))
y2 = np.exp(np.log(x))

plt.plot(x,y1)
plt.plot(x,y2,'o')
plt.axis('square')
```

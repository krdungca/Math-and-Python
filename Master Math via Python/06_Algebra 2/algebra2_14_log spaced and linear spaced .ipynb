Linear functions are good local approximations off nonlinear functions.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.linspace(1,2,50)
```

```python
np.logspace(1,2)#10^start up to 10^end
```

```python
a = np.log10(1)
b = np.log10(2)

np.logspace(a,b,10)
```

```python
#exercise

# pre_set_axis_lim = 0,100

n_start = 1
n_end = 100
n_spacing = 100


#linear-x
x_lin = np.linspace(n_start,n_end,n_spacing)
plt.plot(x_lin,x_lin,'r',linewidth = 5)


#log-x
x_log_start = np.log10(n_start)
x_log_end = np.log10(n_end)

x_log = np.logspace(x_log_start,x_log_end,n_spacing)

plt.plot(x_lin,x_log,'b',linewidth = 5)


#linear-x decreasing boundaries
for i in range(n_start, n_end-1):
    # xval_line_approx = x_lin[i],   x_lin[-(i+1)]
    # yval_line_approx = x_log[i],   x_log[-(i+1)]
    plt.plot([x_lin[i],   x_lin[-(i+1)]],
             [x_log[i],   x_log[-(i+1)]],
             color='green')

axis = plt.gca()
plt.axis('square')
# axis.set_xlim([40,41])


# for i in lin_approx:
#     plt.plot()
# print(x_lin)
# print(x_lin[1])
# print(x_lin[-2])
# print(x_log)

# xval_line_approx = x_lin[i],   x_lin[-(i+1)]
# yval_line_approx = x_log[i],   x_log[-(i+1)]
```

EVEN functions (quadratic style,, can be halved)

ODD function (x**3 style.. )

Formal definition:


$f(x)$ is EVEN --> $f(-x) = f(x)$

$f(x)$ is ODD ---> $f(-x) = - f(x)$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,50)

fEven = x**4
fEvenNeg = (-x)**6

plt.plot(x,fEven,'ro')
plt.plot(x,fEvenNeg,'k-')

fOdd = x**3
fOddNeg = (-x)**3

plt.plot(x,fOdd)
plt.plot(x,fOddNeg)

plt.ylim([-30,30])
```

```python
x = np.linspace(-2*np.pi,2*np.pi,39)

#fx = sin(x)
fx_sin = np.sin(x)
fx_sin_neg = np.sin(-x)
plt.subplot(1,2,1)
plt.plot(x,fx_sin)
plt.plot(x,fx_sin_neg,'ro')


#fx = cos(x)
fx_cos = np.cos(x)
fx_cos_neg = np.cos(-x)
plt.subplot(1,2,2)
plt.plot(x,fx_cos)
plt.plot(x,fx_sin_neg,'ro')

#adjust subplots
plt.subplots_adjust(wspace=0.4)
```

```python
theta = np.linspace(-2*np.pi,2*np.pi,39)

cosfun = np.cos(theta)
cosfunN = np.cos(-theta)
plt.subplot(1,2,1)
plt.plot(theta,cosfun)
plt.plot(theta,cosfunN,'ro')

sinfun = np.sin(theta)
sinfunN = np.sin(-theta)
plt.subplot(1,2,2)
plt.plot(theta,sinfun)
plt.plot(theta,sinfunN,'ro')

#adjust subplots
plt.subplots_adjust(wspace=0.4)
```

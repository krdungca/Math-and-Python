```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lst = [-4,5.5,1,2]

print('minimum: ',np.min(lst))
print('maximum: ',np.max(lst))

maxval = np.max(lst)

maxvalIDX = np.argmax(lst)  #outputs the index of the maximum value

lst[maxvalIDX]
```

```python
x = np.linspace(0,2*np.pi,50)

fx = -(np.cos(x) + np.sqrt(x))

plt.plot(x,fx)

fmax = np.max(fx)
fmaxIDX = np.argmax(fx)

plt.plot(x, fx,'bo')
plt.plot(x[fmaxIDX],fmax,'ro')

print(fmax,fmaxIDX)
```

```python
x = np.linspace(-2,2,51)

fx = x**3 + x**4

fx_min_idx = np.argmin(fx)

plt.plot(x,fx,'bo-')
plt.plot(x[fx_min_idx],fx[fx_min_idx],'ro-')

plt.xlim([-1.2,0.5])
plt.ylim([-.106,-.1])
```

```python
x = np.linspace(-2,2,100)

fx = x**3 + x**4

fx_min = np.min(fx)

fx_min_idx = np.argmin(fx)


plt.subplot(1,2,1)
plt.plot(x[fx_min_idx],fx_min,'ro')

plt.plot(x,fx, 'b-')

plt.subplot(1,2,2)
plt.subplots_adjust(wspace=0.5, hspace=1)
plt.plot(x,fx, 'b-')
plt.plot(x[fx_min_idx],fx_min,'ro')
plt.xlim([-1.2,.5])
plt.ylim([-.2,.2])

```

```python
import scipy
import scipy.signal



x = np.linspace(0,12*np.pi,200)

fx = -np.cos(x) - np.sqrt(x)

plt.plot(x,fx,'b-')

peaks = scipy.signal.find_peaks(fx)  #outputs the position of the PEAKS not the value

troughs = scipy.signal.find_peaks(-fx)

for i in peaks[0]:
    plt.plot(x[i],fx[i],'ro')

for i in troughs[0]:
    plt.plot(x[i],fx[i],'ko')

```

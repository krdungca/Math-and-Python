```python
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5

plt.plot(x,y,'ro')   #r for red, o for circle.. search nalang 

plt.axis('square') #squareplot
plt.grid()

plt.axis([0,6,0,+6])  #input is (x,x,y,y)
```

```python
x = [-4,5,2,3,4,5,3,-5]
y = [0,3,4,5,4,3,2,3]


for i in range(0,len(x)):
    plt.plot(x[i],y[i],'o',label='point %s' %(i+1))

plt.axis('square')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(4,2,'r+')

axis = plt.gca() #gca --- get current axis

ylim = axis.get_ylim() #get_xlim --- limits of x axis


axis.set_ylim(ylim[0],5.8)  #set the new y_lim's 

print(ylim)
```

```python
#exercise
#plot   f(x) = x**2 -3*x,   x range -10 to 11
#add label to axis

import sympy as sym

from sympy.abc import x

fx = x**2 - 3*x
x_range = range(-10,12)

for i in x_range:
    plt.plot(i,fx.subs(x,i),'o')

plt.ylabel('f(x) = %s' %fx)
plt.xlabel('x')
plt.show()
```

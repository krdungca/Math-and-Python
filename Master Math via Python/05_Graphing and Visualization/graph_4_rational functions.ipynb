```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(-3,4)

y = np.zeros(len(x))

for i in range(0,len(x)):
    y[i] = 2 - x[i]**2

plt.plot(x,y,'s-')  #squares with line connecting the squares
```

```python
x = np.linspace(-3,3,2000) #lower bound, upper bound, number of steps b/w bounds
y1 = 2 + (abs(x))**(1/3)
y2 = 2 + (abs(x))**(1/2)

plt.plot(x,y1,'m-')
plt.plot(x,y2,'k-')
```

```python
#plot y = x**(-1 to 3)
import sympy as sym
from IPython.display import display, Math

x = np.linspace(-5,5,200)
x_exp = range(-1,4)

for i in range(0,len(x_exp)):
    y = x**(x_exp[i])
    plt.plot(x,y, linewidth=2.5, label='$y=x^{%s}$' %x_exp[i]) #use $----code in lateX-----$

axis = plt.gca()

axis.set_xlim(x[0],x[-1])
axis.set_ylim(-20,20)
    
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
```

```python
#points of polygons are in [x,y] pair.. not array-like

y = np.array([[1,1],[2,3],[3,1]])  #3points within an array

p = Polygon(y,facecolor='k',alpha=0.8)   #alpha = transparency b/w 1 and 0

fig,ax = plt.subplots()

ax.add_patch(p)

ax.set_ylim([0,4])
ax.set_xlim([0,4])
plt.show()
```

```python
#f(x) = -x**2

x = np.linspace(-2,2,250)

f = -x**2

pnts = np.vstack((x,f)).T
p = Polygon(pnts,facecolor='g',alpha=0.2,edgecolor='k')

y1 = np.array([[-.5,-4],[-0.5,-2.5],[.5,-2.5],[.5,-4]])
p1 = Polygon(y1,facecolor='k')

fig,ax = plt.subplots()

ax.add_patch(p)
ax.add_patch(p1)

plt.plot(x,f,'k')
plt.plot([-2,2],[-4,-4],'k')
plt.axis('off')

```

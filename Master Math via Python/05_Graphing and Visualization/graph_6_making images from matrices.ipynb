```python
# each matrix is assigned a value corresponding to a color
# smaller--> lighter     higher ---> darker


# formula in creating the image matrix

# a_(i,j) = 3*i - 4*j
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
A = [[1,2],[1.3,4]]

plt.imshow(A)  #.imshow  --> image show

plt.xticks([0,1])
plt.yticks([0,1])
```

```python
A = np.zeros((10,14))

for i in range(0,np.shape(A)[0]):  #i for rows
    for j in range(0,np.shape(A)[1]): #j for columns
        A[i,j]=3*i - 4*j

plt.imshow(A)
plt.plot([0,3],[8,2],'r',linewidth=4,)

for i in range(0,np.shape(A)[0]):  #i for rows
    for j in range(0,np.shape(A)[1]): #j for columns
        plt.text(j,i,int(A[i,j]),horizontalalignment='center',verticalalignment='center')
            #x axis corresponds to the column, y axis corresponds to the row
# print(np.shape(A))  #np.shape to determine the SHAPE ('dimensions') of the matrix (row, col)


plt.set_cmap('Purples')
```

```python
#checkerboard
#equation:  c_i,j = -1**(i+j)
n = 10

C = np.zeros((n,n))

for i in range(0,np.shape(C)[0]):
    for j in range(0,np.shape(C)[1]):
        C[i,j] = (-1)**(i+j)

plt.tick_params(labelleft = False, labelbottom=False)

plt.imshow(C,cmap='gray')

plt.show()
```

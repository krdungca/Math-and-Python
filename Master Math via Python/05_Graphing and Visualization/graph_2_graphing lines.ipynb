```python
import matplotlib.pyplot as plt
```

```python
#line needs start point, and end point

p1 = [-3,-1]
p2 = [4,4]

plt.plot([p1[0],p2[0]],[p1[1],p2[1]], color=[.4,.2,.7], linewidth=5)   #plt.plot(x,y)   color=[r,g,b] where r is in terms of percent

plt.axis('square')
plt.axis([-6,6,-6,6])   #(start-x, end-x, start-y,end-y)

plt.show()
```

```python
x = 3
y = 5

plt.plot(x,y,'ro')
plt.plot([0,x],[0,y], 'r')
# plt.grid()
plt.axis('square')
plt.axis([-6,6,-6,6])

plt.plot([-6,6],[0,0],'k')
plt.plot([0,0],[-6,6],'k')
```

```python
#exercise

import sympy as sym

from sympy.abc import x,y

y = sym.sqrt(sym.Abs(x))

x_range = range(-20,21)

for i in x_range:
    y_eval = y.subs(x,i)
    plt.plot([0,i],[0,y_eval])
```

```python
#plot square

#0,0   2,0   2,2    0,2

sqr_points = [[0,0],[2,0],[2,2],[0,2]]

xval_set = []
yval_set = []

for i in sqr_points:
    xval_set.append(i[0])

for j in sqr_points:
    yval_set.append(j[1])

for k in range(0,len(xval_set)):  
    plt.plot([xval_set[k-1],xval_set[k]],[yval_set[k-1],yval_set[k]])

plt.axis('square')
plt.axis([-2,4.5,-3,5])
plt.grid()

```

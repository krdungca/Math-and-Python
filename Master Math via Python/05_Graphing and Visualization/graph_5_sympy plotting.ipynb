```python
import sympy as sym

import sympy.plotting.plot as symplot

import numpy as np

from sympy.abc import x
```

```python
y = x**2

p = symplot(y,show=False)  #python assumes x-values,,,, assumes y=(variable or values)
p.xlim = [0,50]
p[0].line_color = 'r'   #p[0] accessing the elements of sym.plotting.plot()... since it is in a list format

title='this is a nice plot'

p.title = title

p.show()


```

```python
x,a = sym.symbols('x,a')

expr = a/x

n=1
p = symplot(expr.subs(a,n),(x,-5,5),show=False)  #importing custom x or y limits

p[0].label = '$y=%s$' %(sym.latex(expr.subs(a,n)))

#add another function using .extend()
p.extend(symplot(expr.subs(a,2),show=False))   #extends the p[], thus will have a new index value
p[1].label = '$y=%s$' %(sym.latex(expr.subs(a,2)))


p.ylim = [-5,5]
p.legend = True
print(p)
p.show()
```

```python
#graph      y = a / (x**2 - a)
#a = {1,2,3,4}


a,y,x = sym.symbols('a,y,x')

a_set = [1,2,3,4]
expr = a / ((x**2) - a)

y = symplot(expr.subs(a,1),show=False)

for i in range(0,len(a_set)):
    y.extend(symplot(expr.subs(a,a_set[i]),show=False))
    y[i].label = '$%s$' %sym.latex(expr.subs(a,a_set[i]))
       
y[-1].label = ''     
y.legend = True
y.xlim = [-5,5]
y.ylim =[-10,10]
y.show()
```

```python
y = a/ (x**2-a)

p = None
p = symplot(y.subs(a,1),(x,-5,5),show=False)
p[0].label = '$%s$' %sym.latex(y.subs(a,1))

for i in range(2,5):
    p.extend(symplot(y.subs(a,i),(x,-5,5),show=False))
    p[i-1].label = '$%s$' %sym.latex(y.subs(a,i))
    p[i-1].line_color = list(np.random.rand(3))

p.legend = True
p.ylim = [-10,10]
p.show()
```

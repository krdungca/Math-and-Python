```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# y = mx + b


x = [-5,5]
m = -0.3 
b = 0.1

# y = [0,0]

# for i in range(0,len(x)):
#     y[i]= m * x[i] + b

y = m*np.array(x) + b

plt.plot(x,y,label='y=%sx+%s'%(m,b)) #x-coordinates, y-coordinates (per point), including axis label
plt.legend()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()  #access multiple features of the plot 

plt.plot(axis.get_xlim(),[0,0], 'k--')   #creating the base axes
plt.plot([0,0],axis.get_ylim(), 'k--')
```

```python
#plot the ff:
# y = 0.7x + -2
# y = -1.25x + 0.75


x = [-4,4]
m = [0.7,-1.25]
b = [-2,0.75]

y1 = [0,0]
y2 = [0,0]

for i in range(0,len(x)):
    y1[i] = m[0]*x[i] + b[0]

for j in range(0,len(x)):
    y2[j] = m[1]*x[j] + b[1]
    
print(x)
print(y1)

#plot line 1
plt.plot(x,y1,'r', label='y=%sx+%s'%(m[0],b[0]))

#plot line 2
plt.plot(x,y2,'b', label='y=%sx+%s'%(m[1],b[1]) )

#add base axes
plt.xlim(x)  #set x-limit
plt.ylim(y) #set y-limit as xvalues (produce square)

axis = plt.gca() #get current axis
plt.plot(axis.get_xlim(),[0,0],'k--')
plt.plot([0,0],axis.get_ylim(),'k--')

#add function labels/activate legend
plt.legend()

#add grid
plt.grid()
```
